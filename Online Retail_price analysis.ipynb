```python
Europe=["Austria", "Belgium", "Channel Islands", "Cyprus", "Czech Republic", "Denmark", "EIRE", "Finland", "France", "Germany", "Greece", "Iceland", "Italy", "Lithuania", "Malta", "Netherlands", "Norway", "Poland", "Portugal", "Spain", "Sweden","Switzerland"]
Middle_East=["Bahrain", "Israel", "Lebanon", "Saudi Arabia", "United Arab Emirates"]
America = ["Brazil","Canada","USA"]
Asia_Pacific = ["Japan", "Hong Kong","Singapore", "Australia"]
UK = ["United Kingdom"]
Others = ["RSA", "Unspecified","European Community"]
All_Countries = Europe + Middle_East + America + Asia_Pacific + UK + Others
All_Countries_No_UK = Europe + Middle_East + America + Asia_Pacific + Others
```

```python
df_cleansed.columns
```

```python
# Get a dataframe of all itmes' average unitprice in UK
df_UK_mean_unitprice = df_cleansed.groupby(["Country"]).get_group("United Kingdom").groupby("StockCode").mean()["UnitPrice"]
```

```python
df_UK_mean_unitprice
```

```python
# Price_level

price_level = pd.read_csv('outside data/price level.csv')

price_level=price_level.set_index('Country Name')

price_level
```

```python
#create a price dataframe for all countries and all data frame
df_price_matrix = df_UK_mean_unitprice.to_frame().copy()
df_price_matrix.columns = ["United Kingdom"]
uk_factor = price_level.loc["United Kingdom"][0]

for country in All_Countries_No_UK:
    country_factor = price_level.loc[country][0]/uk_factor
    df_price_matrix[country]=df_price_matrix["United Kingdom"]*country_factor
```

```python
df_price_matrix
```

```python
df_price_matrix["USA"][16045]
```

```python
df_price_matrix.to_csv(r'price_matrix.csv', index = True, header=True)
```

```python
df_price_matrix.index
```

```python
new_groupby = df_cleansed.groupby("Country")
country_list = []
for country in All_Countries:
    total_country_revenue = 0
    
    #groupby
    country_groupby = new_groupby.get_group(country)
    
    #stockcode
    stock_array = country_groupby["StockCode"].unique()
    
    #
    for stockcode in stock_array:
        #[0]
        if stockcode in df_price_matrix.index:
            no_of_product = country_groupby[country_groupby["StockCode"]==stockcode]["Quantity"].sum()
            #no_of_product = df_cleansed.groupby(["Country","StockCode"])["Quantity"].sum()[country][stockcode[0]]
            price_of_product = df_price_matrix[country][stockcode]
            total_country_revenue+=no_of_product*price_of_product
        else:
            no_of_product = country_groupby[country_groupby["StockCode"]==stockcode]["Quantity"].sum()
            price_of_product = country_groupby["UnitPrice"].mean()
            total_country_revenue+=no_of_product*price_of_product
    country_list.append([country, total_country_revenue])
        
```

```python
country_list
```

```python
my_df = pd.DataFrame(country_list)

my_df.to_csv('price_matrix_increase_revenue.csv', index=False, header=False)
```
